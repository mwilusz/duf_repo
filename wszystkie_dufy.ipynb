```python
import requests
import pandas as pd
import re
from time import sleep
```

```python

#adres wyszukiwania
url="https://www.ebi.ac.uk/interpro/api/entry/interpro/?search=duf&page_size=200"


DUF_list=[]
short_name_list=[]

while url:
    response=requests.get(url, headers={"Accept": "application/json"})
    data=response.json()

    for entry in data["results"]:
        if "pfam" in entry["metadata"]["member_databases"]:
            if re.search("duf[0-9]+",entry["metadata"]["name"],re.IGNORECASE):
                #element=entry["metadata"]["member_databases"]["pfam"]
                DUF_list.append(list(entry["metadata"]["member_databases"]["pfam"].keys())[0])
                short_name_list.append(re.search("duf[0-9]+",entry["metadata"]["name"],re.IGNORECASE)[0].upper())

    sleep(10)
    # następna strona
    url=data.get("next")


```

```python
print("Accessions:", DUF_list)
print("Short names:", short_name_list)
```

```python
all_duf_names=pd.DataFrame({
    "accession":DUF_list,
    "duf_number":short_name_list
})
```
